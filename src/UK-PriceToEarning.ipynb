```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
price_earnings_ratio = pd.read_csv('../datasets/price_earnings_eng_cym_wrangled.csv')
price_earnings_ratio
```

```python
extract = pd.DataFrame(price_earnings_ratio)
extract = extract[extract['Code'].isin(['E92000001', 'W92000004'])]
df = extract[['Name', 'Year', 'price_to_earning', 'Low_q_price_to_earning']]
```

```python
df = df.sort_values(by=['Name', 'Year']).reset_index(drop=True)
```

```python
plt.figure(figsize=(14,7))
plt.title('Trend of Price to Earning Ratio Over Time (Median)')
plt.xlabel('Year')
plt.ylabel('Price to Earning Ratio')
plt.grid(True)

df['Year'] = pd.to_datetime(df['Year'])
df['Plot_Year'] = df['Year'].dt.year
years = df['Plot_Year'].unique()

for name, group in df.groupby('Name'):
    plt.plot(group['Plot_Year'], group['price_to_earning'], label=name)
    plt.plot(group['Plot_Year'], group['Low_q_price_to_earning'],
             label=f'{name} - Lower 25%', linestyle='--')

plt.legend()
plt.show()
```
