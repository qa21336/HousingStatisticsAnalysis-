```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

completions_df_full = pd.ExcelFile("/content/drive/MyDrive/ADS Datasets/statistic_id746101_housing-completions-in-the-uk-1949-2022-by-tenure.xlsx")
completions_df_data = pd.read_excel(completions_df_full, 'Data')
completions_df_data
```

```python
#Remove unneeded empty columns/rows
completions_df_data.drop(completions_df_data.columns[0], axis=1, inplace=True)
completions_df_data.drop(completions_df_data.index[[0,1,2]], axis=0, inplace=True)
completions_df_data.reset_index(drop = True, inplace= True)
completions_df_data.columns = completions_df_data.iloc[0]
completions_df_data.drop(completions_df_data.index[0], axis=0, inplace=True)
completions_df_data.reset_index(drop = True, inplace= True)
completions_df_data
```

```python
#Add totals
column_list =list(completions_df_data)
print(column_list)
del column_list[0]
print(column_list)
completions_df_data['Total'] = completions_df_data[column_list].sum(axis=1)
completions_df_data
```
