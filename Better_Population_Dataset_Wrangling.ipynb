# New (longer) Population Dataset wrangling

```python
englandPopBySexDf = pd.read_excel("/content/drive/MyDrive/ADS Datasets/ukpopulationestimates18382020.xlsx",
                                  sheet_name="Table 10",
                                  header=4)
walesPopBySexDf = pd.read_excel("/content/drive/MyDrive/ADS Datasets/ukpopulationestimates18382020.xlsx",
                                  sheet_name="Table 12",
                                  header=4)

englandPopBySexDf.head()
```

## England

```python
import numpy as np

def wrangle(df, country):
    dfYearCol = df["Year"]
    newdfYearCol = list(map(lambda y: y.split("-")[1], dfYearCol))
    df["Year"] = newdfYearCol
    df["country"] = np.repeat(country, len(df))
    df = df.rename(columns={"Year": "year"})

    return df

newEnglandPopBySexDf = wrangle(englandPopBySexDf, "E")
newEnglandPopBySexDf.head()
```

## Wales

```python
newWalesPopBySexDf = wrangle(walesPopBySexDf, "W")
newWalesPopBySexDf.head()
```

## Merge England and Wales

```python
popDf = pd.concat((newEnglandPopBySexDf, newWalesPopBySexDf), ignore_index=True)
popDf
```

### Add to Drive
